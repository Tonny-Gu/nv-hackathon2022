```python
import matplotlib.pyplot as plt
import numpy as np
from brokenaxes import brokenaxes
```

```python
data_len = [10**4/2, 10**4, 10**5/2, 10**5, 10**6/2, 10**6, 10**7/2, 10**7, 10**8/2, 10**8]
sorting =   [0.051232, 0.061376, 0.086010, 0.102848, 1.259776, 1.273856, 2.543168, 3.937792, 15.962336, 32.995262]
sp_sort =   [0.047104, 0.049152, 0.047392, 0.051520, 0.086016, 0.077792, 0.107360, 0.113888, 1.224448, 1.390112]
thrust =    [0.135552, 0.130336, 0.130368, 0.147584, 0.714752, 0.809568, 1.037440, 1.366016, 3.010560, 4.994720]
step4 =     [0.007392, 0.006784, 0.007168, 0.007680, 0.019904, 0.034688, 0.136448, 0.264192, 1.291648, 2.569600]
step5 =     [0.006144, 0.006240, 0.006144, 0.006144, 0.016064, 0.027072, 0.106400, 0.205120, 0.997760, 1.977952]
```

```python
fig = plt.figure(figsize = (10,8))
bax = brokenaxes(ylims=((0,6),(28,30)))
bax.set_title("Threshold Seeking Time")
bax.set_xlabel("Data Length", labelpad=35)
bax.set_ylabel("Time(ms)")
bax.plot(data_len,sorting,label="Sorting")
bax.plot(data_len,sp_sort,label="Sample Sorting") 
bax.plot(data_len,thrust,label="Thrust") 
bax.plot(data_len,step4,label="Double Loading") 
bax.plot(data_len,step5,label="Double Loading + Unrolling")
bax.set_xscale('log')
bax.legend(loc = "upper left")
bax.grid()
```
